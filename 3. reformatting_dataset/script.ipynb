```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
path = '/content/drive/MyDrive/praksa_manu/datasets/stock_tweets.csv'

data = pd.read_csv(path)
```

```python
data[:10]
```

```python
data['Date'] = pd.to_datetime(data['Date'])


close_prices = []

for index, row in data.iterrows():
    stock = row['Stock Name']
    date = row['Date'].date()


    try:
        stock_data = yf.download(stock, start=date, end=date + pd.Timedelta(days=1))

        if not stock_data.empty:
            close_price = stock_data.iloc[0]['Close']
        else:
            close_price = None
    except Exception as e:
        print(f"Error fetching data for {stock} on {date}: {e}")
        close_price = None

    close_prices.append(close_price)


data['Close'] = close_prices

```

```python
path = '/content/drive/MyDrive/praksa_manu/reformatting_dataset/reformatted_stock_data.csv'

data.to_csv(path)
```
